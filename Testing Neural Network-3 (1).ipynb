```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt



arr_x= np.random.randn(429,1)   #datatype: numpy array, Dimention- 1D, size- (429,)
arr_y = arr_x+1    # datatype: numpy array, Dimention- 1D, size- (429,)

X_train, X_test, Y_train, Y_test = train_test_split(arr_x, arr_y, test_size=0.3)
#X_train, X_test, Y_train, Y_test = torch.utils.data.random_split(arr_x, arr_y, test_size=0.3)


#train_data = datasets.ImageFolder(data_dir + '/train', transform=train_transforms)
#test_data = datasets.ImageFolder(data_dir + '/test', transform=test_transforms)

#trainloader = torch.utils.data.DataLoader(train_data, batch_size=32)
#testloader = torch.utils.data.DataLoader(test_data, batch_size=32)

#torch.utils.data.random_split

print(arr_x.size)
#print(arr_y)
#print(X_train)
#print(X_test)
#print(Y_test)
#print(Y_train)

A=0
B=0
C=0
D=0

class NeuralNetwork:
    def __init__(self):
        self.inputsize = 1
        self.outputsize = 1
        self.hiddensize = 32
        
    #weights
        self.W1 = np.random.randn(self.inputsize, self.hiddensize)
        self.W2 = np.random.randn(self.hiddensize, self.outputsize)
        
    def forward(self, X):
        #self.z = np.dot(X, self.W1)
        self.z = np.dot(self.W1.T, X.T)
        self.z2 = self.sgmoid(self.z)
        
        #self.z3 = np.dot(self.z2,self.W2)
        self.z3 = np.dot(self.W2.T,self.z2)

        y_predicted = self.sgmoid(self.z3)
        
        return y_predicted
    
    def sgmoid(self, s):
        return 1/(1+np.exp(-s))
    
    def sgmoidPrime(self, s):
        return (s) * (1-(s))
        
    #def sgmoid(self, s):
        #return np.log(1+np.exp(s))
    
    #def sgmoidPrime(self, s):
        #return 1/(1+np.exp(-s))
        
    
    def backward(self, X, y, y_predicted):
        #self.y_predicted_e = y - y_predicted_e                                       #Y^ -Y
        #self.y_pred_delta = self.y_predicted_e * self.sgmoidPrime(y_predicted)       #ypredicted-remove line
        
        self.y_pred_delta = y_predicted - y.T
        
        #B=y_predicted
        #A=y_pred_delta
        
        self.z2_error = self.W2.dot(self.y_pred_delta)
        self.z2_delta = self.z2_error*self.sgmoidPrime(self.z2) 

        self.z2_deltaA = self.W2.T*self.y_pred_delta
        self.z2_delta = (self.z2_deltaA)*(self.sgmoidPrime(self.z))
        
        
        self.W1 += np.mean(X * (self.z2_delta), axis = 0)
        self.W2 += np.mean( self.z2.T*(self.y_pred_delta),axis=1)
        
    def train(self, X, y):
        y_predicted = self.forward(X)
        self.backward(X, y, y_predicted)
        
    def saveWeights(self):
        np.savetxt("w1.txt", self.W1, fmt="%s")
        np.savetxt("w2.txt", self.W2, fmt="%s")
        
    def predict(self):
        print ("Predicted data based on trained weights: ")
        print ("Input (scaled): \n" + str(X_test))
        print ("Output: \n" + str(self.forward(X_test)))
        
        
NN = NeuralNetwork()
for i in range(2): # trains the NN 1,000 times
    #print ("# " + str(i) + "\n")
    #print ("Input (scaled): \n" + str(X))
    #print ("Actual Output: \n" + str(Y_train))
    print("Delta \n" + str(A))
    print("")
    print ("Predicted Output: \n" + str(NN.forward(X_train)))
    #print ("Loss: \n" + str(np.mean(np.square(Y_train - NN.forward(X_train))))) # mean sum squared loss
    #print ("\n")
    NN.train(X_train, Y_train)


#NN.saveWeights()
#NN.predict()

#plt.plot(Y_test)
#plt.show()


```
